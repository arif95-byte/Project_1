One Piece Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the data
data = pd.read_csv('OnePieceArcs.csv')
```

```python
#Display basic information about the dataset
print(data.head())
print(data.info())
```

```python
#Summary statistics of numerical column
print(data.describe())
```

```python
#Plotting distribution of manga completion percentage using a histogram
plt.figure(figsize=(8,6))
plt.hist(data['Manga%'], bins=20, edgecolor='black')
plt.title('Distribution of Manga Completion Percentage')
plt.xlabel('Manga Completion Percentage')
plt.ylabel('Count')
plt.show()
```

```python
#Plotting distribution of anime completion percentage using a histogram
plt.figure(figsize=(8,6))
plt.hist(data['Anime%'], bins=20, edgecolor='black')
plt.title('Distribution of Anime Completion Percentage')
plt.xlabel('Anime Completion Percentage')
plt.ylabel('Count')
plt.show()
```

```python
#Plotting the relationship between manga chapters and anime episodes using a scatter plot
plt.figure(figsize=(8,6))
plt.scatter(data['TotalChapters'], data['TotalEpisodes'], marker='o', alpha=0.7)
plt.title('Manga Chapters vs. Anime Episodes')
plt.xlabel('Total Manga Chapters')
plt.ylabel('Total Anime Episodes')
plt.show()
```

```python
#Plotting the relationship between manga pages and anime minutes using a scatter plot
plt.figure(figsize=(8,6))
plt.scatter(data['TotalPages'], data['TotalMinutes(avg 24)'], marker='o', alpha=0.7)
plt.title('Manga Pages vs. Anime Minutes')
plt.xlabel('Total Manga Pages')
plt.ylabel('Total Anime Minutes')
plt.show()
```
